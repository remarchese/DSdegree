```python
# Essas 65 ações e 65 códigos são da primeira página como faço para pegar de outras páginas tb?

# Pode pegar os links dentro da class "odd select first" ou soup.find_all(class_="link_list atoz_list")[0].get_text()

# criar os links para buscar os dados

# https://br.advfn.com/bolsa-de-valores/bovespa/A'
# https://br.advfn.com/bolsa-de-valores/bovespa/B'
# https://br.advfn.com/bolsa-de-valores/bovespa/C'
# https://br.advfn.com/bolsa-de-valores/bovespa/D'
# https://br.advfn.com/bolsa-de-valores/bovespa/E'
# https://br.advfn.com/bolsa-de-valores/bovespa/F'
# https://br.advfn.com/bolsa-de-valores/bovespa/G'

# https://br.advfn.com/bolsa-de-valores/bovespa/X 

import string # letras do alfabeto

#lista de letras do alfabeto
list(string.ascii_uppercase)
```

```python
 # montando o primeiro link
'https://br.advfn.com/bolsa-de-valores/bovespa/' + list(string.ascii_uppercase)[0]
```

```python
paginas = []
for item in list(string.ascii_uppercase):
    paginas.append('https://br.advfn.com/bolsa-de-valores/bovespa/' + item)

for i in range(10):
    paginas.append('https://br.advfn.com/bolsa-de-valores/bovespa/' + str(i))
```

```python
# páginas para acessar
paginas
```

```python
# criar uma função para buscar info em cada página:

def pagina_ativos(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    acao = []
    codigo = []
    for nome, ticker in zip( soup.find_all(class_='String Column1'), soup.find_all(class_='String Column2 ColumnLast')):
        acao.append(nome.get_text())
        codigo.append(ticker.get_text())
    
    return acao, codigo
    
```

```python
# variável paginas tem a lista de todos os links

# testando a função:
pagina_ativos(paginas[0])
```

```python
# rodar para todas as páginas
lst_page_acao = []
lst_page_codigo = []

for url in paginas:
    print(url)
    acao, codigo = pagina_ativos(url)
    lst_page_acao.extend(acao)
    lst_page_codigo.extend(codigo)
    
    print(len(lst_page_acao), len(lst_page_codigo))

    
```

```python
# print(lst_page_acao)
print(lst_page_codigo)
```

```python
# Bonus 1
# Não foi dado mas em breve vc vão aprender e usar

import pandas as pd

# conda install -c anaconda pandas

df = pd.concat([pd.DataFrame(lst_page_acao, columns=['nome da ação']), 
                pd.DataFrame(lst_page_codigo, columns=['codigo da ação'])], axis=1)
df
```
