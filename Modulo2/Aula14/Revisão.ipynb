```python
# BVSP2 = pd.read_csv('BVSP2.csv', sep=',') 
# OU
# BVSP2 = pd.read_csv('BVSP2.csv') # por padro a separao  por ,
# OU
BVSP2 = pd.read_csv('BVSP2.csv', delimiter=',')
BVSP2
```

```python
# quero trabalhar com a data
import datetime as dt

BVSP2['Date'] = pd.to_datetime(BVSP2['Date']) # transformando a coluna Date de objeto para datetime

BVSP2.info()
```

```python
# extrair o dia 
BVSP2['Date'].dt.day

# extrair s o ms
BVSP2['Date'].dt.month

# extrair o ano
BVSP2['Date'].dt.year
```

```python
# maior valor de abertura
BVSP2['Open'].max()
```

```python
# relao entre Abertura e Fechamento de cada dia desde 01/07/2021

df = BVSP2[BVSP2['Date']>'2021-07-01']
# quando trabalhamos com datas, o ideal para fazer grfico,  colocar como ndice
# fazer o grfico
df.set_index('Date')[['Open','Close']].plot()
```

```python
#boxplot de abertura e fechamento aps 01/07/2021
fig, ax = plt.subplots(2,1, sharex=True)

sns.boxplot(x='Open', data=df, ax=ax[0])
sns.boxplot(x='Close', data=df, ax=ax[1], color='orange')
```

```python
plt.bar(x='day_name',height='High', data=df)
plt.bar(x='day_name',height='Low', data=df, alpha=0.5)
```

```python
plt.stackplot(df['Date'],
              [df['Open'], df['Close']],
              labels=['Open', 'Close'],
              alpha=0.5) # transparncia

plt.legend(loc=2, fontsize='large')
plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
# correlao entre as medidas dos pinguins
penguins.corr()
```

```python
# mapa de calor dessa correlao
sns.heatmap(penguins.corr(), annot=True, cmap='plasma')
```

```python
# distribuio da massa corporal do pinguim
sns.displot(data=penguins, x='body_mass_g', kde=True)
```

```python
# ver a relao de todas as medidas do pinguim em uma nica figura

sns.pairplot(data=penguins, hue='species')
```

```python
# grfico de pizza das espcies de pinguins dos dados
# quantos pinguins de cada espcies

# penguins['species'].value_counts().plot.pie(autopct='%1.1f%%')
# OU
penguins['species'].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

```python
# quero saber a mdia do peso dos pinguins da espcie Adelie e Gentoo de cada ilha

lista = ['Adelie', 'Gentoo']
penguins[penguins['species'].isin(lista)].groupby('island')['body_mass_g'].mean()
```

```python
# penguins[(penguins['species']=='Adelie')|(penguins['species']=='Gentoo')]
# ou
# penguins[penguins['species'].isin(['Adelie', 'Gentoo'])]
```

```python
# grfico de barras horizontais
penguins[penguins['species'].isin(lista)].groupby('island')['body_mass_g'].mean().plot.barh()
```

```python
# ordenar os valores de forma ascendente
penguins[penguins['species'].isin(lista)].groupby('island')['body_mass_g'].mean().sort_values()
```
