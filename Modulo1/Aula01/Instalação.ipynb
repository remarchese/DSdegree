**Jupyter**  
Os notebooks so uma forma de programao literria (documentao seja escrita como uma narrativa junto ao cdigo em vez de ser uma coisa independente)

O ponto central  o servidor do notebook. A conexo  feita no servidor por seu navegador, e o notebook  carregado como um aplicativo web. O cdigo escrito nesse aplicativo  mandado pelo servidor para o ncleo. O ncleo roda o cdigo e o manda de volta para o servidor, ento, o output  carregado no navegador. Ao salvar um notebook, ele  escrito no servidor como um arquivo JSON com a extenso .ipynb. O novo nome, Jupyter, vem da combinao dos nomes Julia, Python e R. Outra vantagem  que o servidor pode ser rodado e acessado via Internet. Normalmente, voc rodar o servidor em sua prpria mquina, onde todos os dados e arquivos notebook esto salvos. Mas,  possvel tambm configurar um servidor em uma mquina remota ou instncia de nuvem, como a EC2, da Amazon. Ento, voc pode acessar os notebooks em um navegador de qualquer lugar do planeta.


conda install jupyter notebook  
jupyter notebook           #lanar um servidor notebook e os arquivos sero salvos nesse diretrio . A prtica padro  inicializar o servidor no diretrio onde os notebooks se encontram.
Ele vai abrir em http://localhost:8888 onde localhost significa o seu computador e 8888  a porta que o servidor est usando ncleos (kernels). No meu caso, estou rodando o servidor em um ambiente Python 3, ento, tenho um ncleo Python 3 disponvel.
Tambm  possvel desligar o servidor inteiro ao pressionar as teclas control + C duas vezes no terminal. De novo, isso desligar imediatamente todos os notebooks, ento, tenha certeza de que tudo est salvo!
O pequeno teclado  a paleta de comandos. Ao clicar nela, aparecer um painel com uma barra de pesquisa em que voc pode procurar por diversos comandos
